```python
# Información completa del sistema
import psutil, os

# CPU
print(f"CPU cores físicos: {psutil.cpu_count(logical=False)}")
print(f"CPU cores lógicos (threads): {psutil.cpu_count(logical=True)}")
print(f"RAM total: {psutil.virtual_memory().total / 1e9:.1f} GB")
print(f"RAM disponible: {psutil.virtual_memory().available / 1e9:.1f} GB")
```
